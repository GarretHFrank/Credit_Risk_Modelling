### Preprocessing Discrete Dummy Variables

```python
# Create a function to find Weight of Evidence (WoE) and Information Value(IV) 
def woe_discrete(df, discrete_variable_name, good_bad_variable_df):
    
    # Set up DataFrame for analysis
    df = pd.concat([df[discrete_variable_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    
    # Create columns for WoE and IV calculations
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()   # calculates the proportion of observations
    df['n_good'] = df['prop_good'] * df['n_obs']         # calculates the number of good (1) loans
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']    # calculates the number of bad (0) loans
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()# calculates the proportion of good (1) loans
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()   # calculates the proportion of bad (0) loans
    
    # Calculates the WoE
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    df = df.sort_values(['WoE'])
    df = df.reset_index(drop = True)
    
    # Calculates the difference between subsequent proportion of good loans and WoE
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    
    # Calcuclate the IV
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    
    return df
```

```python
# Function to visualize WoE 

def plot_by_woe(df_WoE, rotation_of_x_axis_varaibles = 0):
    
    # Establish x and y axis variable
    x = np.array(df_WoE.iloc[: , 0].apply(str))
    y = df_WoE.loc[:, "WoE"]
    
    # Set fig size
    plt.figure(figsize = (18,6))
    
    plt.plot(x,y)
    plt.plot(x,y,marker = "o", linestyle = "--", color = "k")
    plt.xlabel(df_WoE.columns[0])
    plt.ylabel("Weight of Evidence")
    plt.title(str("Weight of Evidence by ") + df_WoE.columns[0])
    plt.xticks(rotation = rotation_of_x_axis_varaibles )
    
```

```python
# Create a function to find Weight of Evidence (WoE) and Information Value(IV) 
def woe_ordered_continuous(df, discrete_variable_name, good_bad_variable_df):
    
    # Set up DataFrame for analysis
    df = pd.concat([df[discrete_variable_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    
    # Create columns for WoE and IV calculations
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()   # calculates the proportion of observations
    df['n_good'] = df['prop_good'] * df['n_obs']         # calculates the number of good (1) loans
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']    # calculates the number of bad (0) loans
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()# calculates the proportion of good (1) loans
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()   # calculates the proportion of bad (0) loans
    
    # Calculates the WoE
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    # Categories are ordered by independent variables
    #df = df.sort_values(['WoE'])
    #df = df.reset_index(drop = True)
    
    # Calculates the difference between subsequent proportion of good loans and WoE
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    
    # Calcuclate the IV
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    
    return df
```
